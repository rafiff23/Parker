```python
import pandas as pd 
```

```python
df = pd.read_csv('leads_clean.csv')
```

```python
df
```

```python
df = df.rename(columns = {'confirmed/not confirmed' : 'Status', '  Fee' : 'Fee'})
df.drop(columns='No', axis=1, inplace=True)
```

```python
df['Fee'] = df['Fee'].str.replace('Rp', '', regex=True)
df['Fee'] = df['Fee'].str.strip()
df['Fee'] = df['Fee'].str.replace(',', '', regex=True)
df['Fee'].fillna(0,inplace=True)
df['Fee'] = df['Fee'].astype('int64')
```

```python
count_value = df[df['Status'] == 'Confirmed']['Fee'].sum()
print(count_value)
```

```python
df_count_bar = df.groupby(["Status Level", "Status"]).size().reset_index(name="Count")
df_count_bar
```

```python
df1 = df.groupby('Status')['Fee'].sum().reset_index().sort_values(by = 'Fee', ascending=False)
```

```python
df1
```

```python
import altair as alt

alt.Chart(df1).mark_bar().encode(
    x=alt.X('Status', sort='-y'),
    y='Fee'
)
```

```python
df2 = df.groupby('Status')['Fee'].count().reset_index().sort_values(by = 'Fee', ascending=False)
```

```python
df2.rename(columns={'Fee' : 'Total Count'}, inplace=True)
```

```python
import altair as alt

chart = alt.Chart(df2).mark_arc(innerRadius=50).encode(
    theta="Total Count:Q",
    color="Status:N",
    tooltip=["Status", "Total Count"]
).interactive()

chart
```
